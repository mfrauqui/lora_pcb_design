```python
import matplotlib.pyplot as plt
temp = [23,12,34,56,23,56]
dew = [1,2,56,23,12,20]
t = range(6)
print(temp)
plt.plot(t,temp,'r')
plt.plot(t,dew,'b')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.show()
```

```python
import matplotlib.pyplot as plt
temp = [23,12,34,56,23,56]
dew = [1,2,56,23,12,20]
t = range(6)
print(temp)
plt.axes([0.05,0,0.425,0.9])
plt.plot(t,temp,'r')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.axes([1,0,0.425,0.9])
plt.plot(t,dew,'b')
plt.xlabel('Dew')
plt.ylabel('Temperature')
plt.show()
```

syntax: subplot(numb_rows, numb_cols, which_subplot_to_activate)

```python
import matplotlib.pyplot as plt
temp = [23,12,34,56,23,56]
dew = [1,2,56,23,12,20]
t = range(6)
print(temp)
plt.subplot(2,1,1)
plt.plot(t,temp,'r')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.subplot(2,1,2)
plt.plot(t,dew,'b')
plt.xlabel('Dew')
plt.ylabel('Temperature')
plt.show()
```

```python
import matplotlib.pyplot as plt
temp = [23,12,34,56,23,56]
dew = [1,2,56,23,12,20]
t = range(6)
print(temp)
plt.subplot(2,1,1)
plt.plot(t,temp,'r')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.subplot(2,1,2)
plt.plot(t,dew,'b')
plt.xlabel('Dew')
plt.ylabel('Temperature')
# Use plt.tight_layout() to improve the spacing between subplots
plt.tight_layout()
plt.show()
```
